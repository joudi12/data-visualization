```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# First we will compare tip VS total bill

```python
sns.relplot(x='tip', y='total_bill', hue=tips.tip.tolist(), data=tips);
```

```python
sns.lmplot(x='tip', y='total_bill', data=tips);
```

# Now let's see who pay more tip

```python
tips_gender = tips.groupby('sex', as_index=False).sum()
tips_gender
```

```python
sns.barplot(x='sex', y='tip', data=tips_gender);
```

# now let's see when the tip increasing in which day 

```python
tips_day = tips.groupby('day', as_index=False).sum()
tips_day
```

```python
sns.barplot(x='day', y='tip', data=tips_day, ci = None);
```
