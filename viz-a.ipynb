```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
car_crashes = sns.load_dataset('car_crashes')
```

```python
car_crashes.head()
```

# How the alcohol effect on the car crashes

### we will compare alchohol VS total because when one increase , the other will increase too
### so lets see this relationship  using  plots to see the effects of alcohol on the total

```python
sns.relplot(x='alcohol', y='total', hue=car_crashes.alcohol, data=car_crashes);
```

sns.lmplot(x='alcohol', y='total',  data=car_crashes)


# let's take the avarage of alcohol affected on car craches 

```python
car_mean =car_crashes.groupby('alcohol', as_index=False).mean()
car_mean
```

```python
sns.lmplot(x='alcohol', y='total', data=car_mean);
```

## let's draw the data in parblot way to see 

```python
graph = sns.barplot(x='alcohol', y='total', data=car_crashes, ci=None)
graph.set_xticklabels(labels=list("")[1:]);
```

## And for the average

```python
graph = sns.barplot(x='alcohol', y='total', data=car_mean,ci=None);
graph.set_xticklabels(labels=list("")[1:]);
```
